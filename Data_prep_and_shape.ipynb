## Practice: Preparing and Shaping Data (Titanic Dataset)

In this practice, we will prepare and shape the Titanic dataset step by step for machine learning.  
We’ll cover:  
- Importing libraries & dataset  
- Exploring data  
- Feature engineering  
- Handling missing values  
- One-hot encoding categorical variables  
- Scaling features  




```python
# ----------------------------------------------------
# Step 1: Import required libraries
# ----------------------------------------------------
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```



---


### Step 2: Load dataset
We use the Titanic dataset (`titanic.csv`) and check the first few rows.


---



```python
data = pd.read_csv('/content/titanic.csv')
```

```python
data.head()
```



---
### Step 3: Explore dataset  
We check summary statistics and data info to understand structure, data types, and missing values.


---




```python
data.describe()
```

```python
data.info()
```



---

### Step 4: Feature Engineering  
We create a new feature **Travelalone**  
- `1` if the passenger is traveling alone  
- `0` if traveling with family


---



```python
data['Travelalone'] = np.where((data['SibSp']+data['Parch'])>0,0,1).astype('uint8')
```



---

### Step 5: Drop unnecessary columns  
These columns don’t help the model (PassengerId, Name, Ticket, Cabin, SibSp, Parch).


---



```python
data1 = data.drop(['PassengerId','Name','Ticket','Cabin','SibSp','Parch'], axis = 1)
```



---
### Step 6: Handle missing values  
Check missing values first.


---




```python
data1.isna().sum()
```



---

We see missing values in **Age**.  
We fill them with the **median** of Age.


---



```python
data['Age'].fillna(data1['Age'].median(skipna = True), inplace = True)
```

```python
data1.head()
```



---

### Step 7: One-Hot Encoding (Convert categorical → numeric)  
We use `pd.get_dummies` for:  
- Pclass  
- Sex  
- Embarked  
(drop_first=True to avoid dummy variable trap)


---



```python
data_titanic = pd.get_dummies(data1, columns = ['Pclass', 'Sex','Embarked'], drop_first = True)
```

```python
data_titanic.head()
```



---
### Step 8: Split features (X) and target (y)  
- **X** → features (independent variables)  
- **y** → target (Survived column)


---




```python
X = data_titanic.drop(['Survived'], axis = 1)
y = data_titanic['Survived']
```



---

### Step 9: Feature Scaling  
We scale numerical features to normalize them.  
- MinMaxScaler → Scales values between [0,1]  
- StandardScaler → Standardizes values (mean=0, std=1)


---



```python
trans_MM = MinMaxScaler()
trans_SS = StandardScaler()
```

```python
data_MM = trans_MM.fit_transform(X)
pd.DataFrame(data_MM)
```

```python
data_SS = trans_SS.fit_transform(X)
pd.DataFrame(data_SS)
```
